```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
	return 3 * x**2 - 4 * x + 5
```

```python
f(5)
```

```python
xs = np.arange(-3, 3, 0.25)
ys = f(xs)

plt.plot(xs, ys)
```

```python
h = 0.0000001
x = 2/3
m = (f(x+h) - f(x))/h
c = f(x) - m*x

ys_1 = m*xs + c
plt.plot(xs, ys)
plt.plot(xs, ys_1)
```
